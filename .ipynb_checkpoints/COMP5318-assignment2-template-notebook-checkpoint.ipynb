#### 1.2 Functions for data preprocessing 

The four functions in the following cell are about:
1. Image normalization
2. PCA
3. Add channel to match the input shape of the NN
4. plot cumulative explained varience

```python
def image_preprocessing(x_train, x_test):
    return x_train / 255, x_test / 255

def apply_pca(x_train, x_test, n_components=0.95):
    x_train = x_train.reshape(-1, 28*28)
    x_test = x_test.reshape(-1, 28*28)
    pca = PCA(n_components=n_components)
    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)
    return x_train_pca, x_test_pca
    
def expand_dimension(x_train,x_val,x_test):
    x_train = np.expand_dims(x_train, -1)
    x_val = np.expand_dims(x_val, -1)
    x_test = np.expand_dims(x_test, -1)
    return x_train, x_val, x_test

def plot_cumulative_explained_variance(x_train_normalized):
    x_train_normalized = x_train_normalized.reshape(-1, 28*28)
    plt.figure(figsize=(8, 5))
    pca_model = PCA()
    pca_model.fit_transform(x_train_normalized)
    explained_variance = np.cumsum(pca_model.explained_variance_ratio_)

    # Determine the number of components needed to achieve 95% variance
    # np.argmax returns zero-based index, thus plus one on it
    num_components = np.argmax(explained_variance >= 0.95) + 1  

    plt.plot(explained_variance, label='Cumulative Explained Variance')

    plt.scatter(num_components - 1, explained_variance[num_components - 1], color='red')  # mark the 95% threshold point here 
    plt.axhline(y=0.95, color='r', linestyle='--', label='95% Explained Variance')
    plt.axvline(x=num_components - 1, color='r', linestyle='--')

    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('Cumulative Explained Variance versus Number of Components')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```

####  In the upcoming cell, we will apply the data preprocessing function to the dataset, 

please notice:
1. PCA and feature flatten are not applied for MLP&CNN
2. train_test_split is used only for MLP&CNN ( SVM uses cv=5 in grid-search later to spilt data)

```python
# Normalization to range [0,1] (for all three algorithms)
X_train_normalized, X_test_normalized = image_preprocessing(X_train, X_test)

# Apply Dimensionality reduction using PCA (for data related to SVM)
X_train_pca, X_test_pca = apply_pca(X_train_normalized, X_test_normalized)

# train and test spilt for NN
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train_normalized, y_train, stratify=y_train, train_size=0.8)

# Input data for neural network 
X_train_split, X_val, X_test_normalized = expand_dimension(X_train_split, X_val, X_test_normalized)

# plot cumulative variance
plot_cumulative_explained_variance(X_train_normalized)
print(X_train_split.shape, X_val.shape, X_test_normalized.shape)
```

### 2.2 Fully connected neural network - without tuning

```python
def build_mlp_model(num_layers, units, activation, lr, optimizer_name):
    model = keras.Sequential()
    model.add(keras.layers.Input(shape=(28,28,1)))
    model.add(keras.layers.Flatten())

    for _ in range(num_layers):
        model.add(
            keras.layers.Dense(
                units=units,
                activation=activation,
            )
        )
    
    model.add(keras.layers.Dense(11, activation="softmax"))

    if optimizer_name == 'adam':
            optimizer = keras.optimizers.Adam(learning_rate=lr)
    elif optimizer_name == 'sgd':
            optimizer=keras.optimizers.SGD(learning_rate=lr)

    model.compile(
        optimizer=optimizer,
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model
```

### 2.3 Convolutional neural network - without tuning

```python
def build_cnn(filter_1, filter_2, kernel_size, lr):
    model = keras.Sequential()
    model.add(keras.layers.Input(shape=(28, 28, 1)))
        # First block
    model.add(keras.layers.Conv2D(filters=filter_1, kernel_size=kernel_size, activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
        # Second block
    model.add(keras.layers.Conv2D(filters=filter_2, kernel_size=kernel_size, activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
        
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(units=64))
    model.add(keras.layers.Dropout(0.2))
    model.add(keras.layers.Dense(11, activation='softmax'))

        # Compile model
    optimizer=keras.optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=optimizer,
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

    return model
```

## 3. Hyperparameter tuning

### 3.1 Algorithm of choice from first six weeks of course

- Support Vector Machine

```python
X_train_pca = X_train_pca[:10000,:]
y_train = y_train[:10000]


# Notice that the initial parameter grid was very big, but it has been significantly narrowed down in this final version after many attempts 
param_grid = [
    {'kernel': ['poly'], 'C': [0.1, 1, 10], 'gamma': [0.005, 0.01], 'degree': [2, 3, 4]},
    {'kernel': ['rbf'], 'C': [0.1, 1, 10], 'gamma': [0.005, 0.01]}
]

svm_tuning = svm.SVC(class_weight='balanced')

grid_search = GridSearchCV(svm_tuning, param_grid, cv=5, scoring='accuracy', verbose=2)
grid_search.fit(X_train_pca, y_train)

print("Best parameters combination for SVM:", grid_search.best_params_)
print("Best cross-validation score for SVM: {:.2f}".format(grid_search.best_score_))
```

```python
results = grid_search.cv_results_
mean_scores = results['mean_test_score']
mean_fit_time = results['mean_fit_time']

# Create a parameter configuration list for latter plotting 
params = []
for i in range(len(results['params'])):
    param_str = str(results['params'][i]).replace("{", "").replace("}", "").replace("'", "").replace(":", "=")
    # Splitting the parameters descriptions to display them in two lines 
    mid = len(param_str) // 2
    break_point = param_str.find(',', mid)  
    if break_point != -1:
        param_str = param_str[:break_point] + '\n' + param_str[break_point+1:]  # Insert newline character
    params.append(param_str)

# Plotting the accuracy and runtime for differnent combinations of hyperparameters
fig, ax1 = plt.subplots(figsize=(7.5, 8.5))  
color = 'tab:blue'
ax1.set_ylabel('Parameter Configuration', fontsize=10)
ax1.set_xlabel('Accuracy', color=color)
ax1.barh(params, mean_scores, color=color, label='Accuracy', height=0.6)
ax1.tick_params(axis='x', labelcolor=color)
ax1.set_yticklabels(params, fontsize=8, va='center')
ax1.legend(loc='upper right')
ax2 = ax1.twiny() 
color = 'tab:red'
ax2.set_xlabel('Runtime (s)', color=color)
ax2.plot(mean_fit_time, params, color=color, marker='o', linestyle='-', label='Runtime')
ax2.tick_params(axis='x', labelcolor=color)
ax2.legend(loc='lower right')
fig.tight_layout()  
plt.show()
```
